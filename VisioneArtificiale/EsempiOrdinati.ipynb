```python
import cv2 as cv
import numpy as np
from va import show

img1 = np.zeros((15, 16), dtype=np.uint8) # Tutti i pixel a 0
img2 = np.full((15, 16), 255, dtype=np.uint8) # Tutti i pixel a 255
img3 = np.random.randint(0, 256, (15, 16), dtype=np.uint8) # Valori dei pixel casuali
img4 = cv.imread('Immagini/building.jpg')

show((img1, "img1"), (img2, "img2"), (img3, "img3"), (img4, "img4"))

for k in range(2,6):
    img5 = img4.copy()
    img5[::k, ::k] = 0
    show((img5, f"Immagine k"))
```

```python
d = np.zeros((15, 16), dtype=np.uint8)
# Integer array indexing e broadcasting
y = np.array([0, 2, 5, 9, 14])
x = np.arange(0, 16, 2)
z = np.arange(50, 255, 50)[:,np.newaxis]
d[y[:, np.newaxis], x] = z

print("Indici x : " + str(x))
print("Indici y : " + str(y))
print("Indici z : " + str(z))
show((d, "1"), (d.copy(), "2"))
```

```python
d = np.random.randint(0, 256, (64, 64, 3), dtype=np.uint8) # Valori dei pixel casuali

dr = d.copy()
dr[..., :2] = 0
db = d.copy()
db[..., 1::] = 0
dg = d.copy()
dg[..., ::2] = 0

d_circle = d.copy()
show((d, "totale"),(dr, "rossa"),(dg, "verde"), (db, "blu"))
show((d_circle, "cerchio"))
```
